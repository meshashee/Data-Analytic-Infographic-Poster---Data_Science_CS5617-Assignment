### Imports

```python
import pandas as pd
```

### Reading & variable initialization

```python
by_sev_1 = 'road_accident_data_by_severity_of_injuries_2010.csv'
by_sev_2 = 'road_accident_data_by_severity_of_injuries_2011.csv'
by_sev_3 = 'road_accident_data_by_severity_of_injuries_2012.csv'
by_vehicle_1 = 'road_accident_data_by_vehicle_type_2010.csv'
by_vehicle_2 = 'road_accident_data_by_vehicle_type_2011.csv'
by_vehicle_3 = 'road_accident_data_by_vehicle_type_2012.csv'

by_sev = [[by_sev_1, '2010'], [by_sev_2, '2011'], [by_sev_2, '2012']]
by_vehicle = [[by_vehicle_1, '2010'], [by_vehicle_2, '2011'], [by_vehicle_3, '2012']]

df_sev = pd.DataFrame()
df_vehicle = pd.DataFrame()
```

### Concat different years of data into a one data set

```python
for i in by_sev:
    df = pd.read_csv(i[0])
    df['Year'] = i[1]
    df_sev = pd.concat([df_sev, df], axis=0)
    df_sev.reset_index(drop=True, inplace=True)

for i in by_vehicle:
    df = pd.read_csv(i[0])
    df['Year'] = i[1]
    df_vehicle = pd.concat([df_sev, df], axis=0)
    df_vehicle.reset_index(drop=True, inplace=True)

df_vehicle.head()
```
